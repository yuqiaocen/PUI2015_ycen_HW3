### 1. Generate

```python
import pylab as pl
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%pylab inline
```

```python
np.random.seed(seed=6)
```

```python
mymean = 100
# the population mean is 100.
sizes = range(10, 2000, 100)
# Generate 100 samples of different size N, 10<N<2000.
sizes = 2000/(np.array(range(1,90))+10)
print (sizes)
```

```python
# The distributions I use.
distributions = {}
distributions ['normal'] = {}
distributions ['poisson'] = {}
distributions ['binomial'] = {}
distributions ['chisq'] = {}
distributions ['gamma'] = {}
distributions ['exponential'] = {}
print (distributions)
```

```python
for n in sizes:
    distributions ['normal'][n] = np.random.normal(mymean, 50, n)
    distributions ['poisson'][n] = np.random.poisson(mymean, n)
    distributions ['binomial'][n] = np.random.binomial(mymean/0.5, 0.5, n)
    distributions ['chisq'][n] = np.random.chisquare(mymean, n)
    distributions ['gamma'][n] = np.random.gamma(50, 2, n)
    distributions ['exponential'][n] = np.random.exponential(mymean, n)
```

### 2. Plot the sample mean against the sample size.

```python
means = {}
means['normal'] = {}
fig_mu_N = pl.figure(figsize=(10, 10))
axnormal_mu_n = fig_mu_N.add_subplot(111)
for nn in distributions['normal'].iterkeys():
    means['normal'][nn] = distributions ['normal'][nn].mean()
    axnormal_mu_n.plot(nn, distributions['normal'][nn].mean(), 'o')
    axnormal_mu_n.set_xlabel('sample size')
    axnormal_mu_n.set_ylabel('sample mean')
    axnormal_mu_n.set_title('Normal Distribution')
means['poisson'] = {}
fig_mu_N = pl.figure(figsize=(10, 10))
axpoisson_mu_n = fig_mu_N.add_subplot(111)
for nn in distributions['poisson'].iterkeys():
    means['poisson'][nn] = distributions ['poisson'][nn].mean()
    axpoisson_mu_n.plot(nn, distributions['poisson'][nn].mean(), 'o')
    axpoisson_mu_n.set_xlabel('sample size')
    axpoisson_mu_n.set_ylabel('sample mean')
    axpoisson_mu_n.set_title('Poisson Distribution')
means['binomial'] = {}
fig_mu_N = pl.figure(figsize=(10, 10))
axbinomial_mu_n = fig_mu_N.add_subplot(111)
for nn in distributions['binomial'].iterkeys():
    means['binomial'][nn] = distributions ['binomial'][nn].mean()
    axbinomial_mu_n.plot(nn, distributions['binomial'][nn].mean(), 'o')
    axbinomial_mu_n.set_xlabel('sample size')
    axbinomial_mu_n.set_ylabel('sample mean')
    axbinomial_mu_n.set_title('Binomial Distribution')
```

```python
means['chisq'] = {}
fig_mu_N = pl.figure(figsize=(10, 10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in distributions['chisq'].iterkeys():
    means['chisq'][nn] = distributions ['chisq'][nn].mean()
    axchisq_mu_n.plot(nn, distributions['chisq'][nn].mean(), 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.set_title('Chisquare Distribution')
means['gamma'] = {}
fig_mu_N = pl.figure(figsize=(10, 10))
axgamma_mu_n = fig_mu_N.add_subplot(111)
for nn in distributions['gamma'].iterkeys():
    means['gamma'][nn] = distributions ['gamma'][nn].mean()
    axgamma_mu_n.plot(nn, distributions['gamma'][nn].mean(), 'o')
    axgamma_mu_n.set_xlabel('sample size')
    axgamma_mu_n.set_ylabel('sample mean')
    axgamma_mu_n.set_title('Gamma Distribution')
means['exponential'] = {}
fig_mu_N = pl.figure(figsize=(10, 10))
axexponential_mu_n = fig_mu_N.add_subplot(111)
for nn in distributions['exponential'].iterkeys():
    means['exponential'][nn] = distributions ['exponential'][nn].mean()
    axexponential_mu_n.plot(nn, distributions['exponential'][nn].mean(), 'o')
    axexponential_mu_n.set_xlabel('sample size')
    axexponential_mu_n.set_ylabel('sample mean')
    axexponential_mu_n.set_title('Exponential Distribution')
```

```python
alpha = 0.05
#we like fractions better then percentages. as a rule of thumb, either use fractions or counts
P_0 = 52.1*0.01 
P_1 = 53.3*0.01

n_0 = 409
n_1 = 564

#lets get the counts by multiplying by the sample size
Nt_0 = P_0*n_0
Nt_1 = P_1*n_1
```

### Z Test

```python
#define the sample proportion first
sp = (P_0*n_0+P_1*n_1)/(n_1+n_0)
print (sp)
```

```python
# i am goonna create a little one line function to calculate the standard dev, it is not really needed, but just to show you how you do such a thing
sp_stdev = lambda p, n: np.sqrt( p * ( 1 - p ) / n[0] +  p * ( 1 - p ) / n[1]  )

sp_stdev_2y = sp_stdev((Nt_0 + Nt_1)/(n_0 + n_1),[n_0,n_1])
print (P_0, n_0, n_1, sp_stdev_2y)
```

### z score

```python
zscore = lambda p0, p1, s : (p0-p1)/s
z_2y = zscore(P_1, P_0, sp_stdev_2y)
print (z_2y)
```
